```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.metrics import roc_auc_score
import pandas as pd
import numpy as np
```

```python
CREDITCARD_CSV = '../lesson-04/creditcard.csv'
df = pd.read_csv(CREDITCARD_CSV)
```

```python
column_names_without_Class = df.columns[0:-1]
X = df[column_names_without_Class]
X
```

```python
y = df['Class']
type(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=100, stratify=y)
print(X_train.shape, X_test.shape)
print(y_train.shape, y_test.shape)
```

```python
parameters = [{
    'n_estimators': [10, 15],
    'max_features': np.arange(3, 5),
    'max_depth': np.arange(4, 7)
}]
```

```python
# https://stackoverflow.com/questions/30102973/how-to-get-best-estimator-on-gridsearchcv-random-forest-classifier-scikit
estimator = RandomForestClassifier(random_state=100)
CV_rfc = GridSearchCV(
    estimator=estimator, param_grid=parameters,
    scoring='roc_auc',
    cv=3)
CV_rfc.fit(X_train, y_train)
CV_rfc.best_params_
```

```python
y_pred_proba = CV_rfc.predict_proba(X_test)
len(y_pred_proba)
#type(y_pred_proba)
```

```python
# вероятность класса 1
y_pred_proba = y_pred_proba.flatten()[1::2]
len(y_pred_proba)
```

```python
roc_auc_score(y_test.array, y_pred_proba)
#type(y_test)
#y_test.array
```
